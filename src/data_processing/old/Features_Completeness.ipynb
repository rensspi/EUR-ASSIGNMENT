```python
df_actueel = df.query("GDB_TO_DATE == '9999-12-31T23:59:59'")

df_boolean = df_actueel.isnull()
```

```python
def return_percentage_true_column(df, column):
    sum_boolean = df[column].value_counts().to_dict()

    percentage_true = 'jammer dit'
    if True in sum_boolean.keys() and False in sum_boolean.keys():
        percentage_true = sum_boolean[True]/(sum_boolean[True]+sum_boolean[False])
    elif True in sum_boolean.keys() and False not in sum_boolean.keys():
        percentage_true = 1
    elif True not in sum_boolean.keys() and False in sum_boolean.keys():
        percentage_true = 0
    return percentage_true

#Percentage True/False per collumn

#let op! de trues zijn dus null-waarden
def return_percentage_true_column_df(df):
    column_dict = {}
    for column in df:
        column_dict[column] = return_percentage_true_column(df, column)

    return column_dict

nulls_per_column = return_percentage_true_column_df(df_boolean)

nulls_per_column



```

```python
def return_percentage_true_row(df):
    nr_columns = len(df.columns)
    sum_rows = df.sum(axis=1).to_dict()
    sum_rows_perc = {}

    for i in sum_rows:
        sum_rows_perc[i] = sum_rows[i]/nr_columns
    
    return sum_rows_perc

nulls_per_row = return_percentage_true_row(df_boolean)

nulls_per_row
```
