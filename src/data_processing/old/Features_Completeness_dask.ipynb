```python

def total_percentage_trues(df):
    total_trues = df.sum().sum().compute()

    nr_rows = df.shape[0].compute()
    nr_cols = len(df.columns)
    total_values = nr_rows * nr_cols

    true_ratio = total_trues / total_values
    return true_ratio

# Som over alle kolommen en rijen
#total_trues = df_boolean.sum().sum().compute()
total_percentage_trues(df_boolean)
#print(total_trues)
```

```python
def return_percentage_values_column(df, column):
    sum_boolean = df[column].value_counts(dropna=False, split_out=8).compute()
    counts_dict = sum_boolean.to_dict()

    freq_series = sum_boolean / sum_boolean.sum()
    freq_dict = freq_series.to_dict()

    return freq_dict

#Percentage True/False per collumn

#let op! de trues zijn dus null-waarden
def return_percentage_true_column_df(df):
    column_dict = {}
    for column in df:
        column_dict[column] = return_percentage_values_column(df, column)

    return column_dict

nulls_per_column = return_percentage_true_column_df(df_boolean)




```

```python
def return_percentage_true_row(df):
    nr_columns = len(df.columns)
    sum_rows_series = (df == True).sum(axis=1).compute()
    sum_rows = sum_rows_series.to_dict()
    sum_rows_perc = {}

    for i in sum_rows:
        sum_rows_perc[i] = sum_rows[i]/nr_columns
    
    return sum_rows_perc

nulls_per_row = return_percentage_true_row(df_boolean)

nulls_per_row
```

```python
def return_percentage_true_row_OID(df):
    """
    Return a dictionary {OBJECTID: percentage_true_per_row} voor een Dask DataFrame
    met kolom 'OBJECTID' en verder alleen booleans.
    """
    # tel True's per rij over alle booleankolommen (OBJECTID niet meenemen)
    sum_rows_series = df.drop(columns=['OBJECTID']).eq(True).sum(axis=1).compute()
    
    # aantal booleankolommen (exclusief OBJECTID)
    nr_columns = len(df.columns) - 1

    # OBJECTID-keys naar geheugen
    objectids = df['OBJECTID'].astype('int64').compute()

    # bouw dictionary {OBJECTID: percentage}
    percentages = (sum_rows_series / float(nr_columns)).values
    return dict(zip(objectids.tolist(), percentages))
```

```python

import numpy as np
import matplotlib.pyplot as plt
from typing import Optional, Tuple, Dict

def histogram_from_pct_dict(
    d: Dict[object, float],
    bins: int = 20,
    percentage_scale: str = "auto",  # '0-1', '0-100', of 'auto'
    value_range: Optional[Tuple[float, float]] = None,  # standaard op basis van schaal
    save_path: Optional[str] = None,   # bv. "hist_percentages.png"
    show: bool = False
) -> Tuple[np.ndarray, np.ndarray]:
    """
    Maakt een histogram van de percentages in d en (optioneel) een plot.
    
    Parameters
    ----------
    d : Dict[object, float]
        Dictionary met {rij_index: percentage}. Percentage kan in 0–1 of 0–100 zitten.
    bins : int
        Aantal bins voor het histogram.
    percentage_scale : str
        'auto' (detecteer 0–1 vs 0–100), of forceer '0-1' / '0-100'.
    value_range : Optional[Tuple[float, float]]
        Bereik van de histogramwaarden. Default hangt af van percentage_scale (0–1).
    save_path : Optional[str]
        Pad om de plot op te slaan. None = niet opslaan.
    show : bool
        True = plot tonen via plt.show(). Handig bij interactief gebruik.

    Returns
    -------
    Tuple[np.ndarray, np.ndarray]
        (counts, bin_edges) zoals geretourneerd door numpy.histogram.
    """
    # 1) Values snel naar NumPy
    vals = np.fromiter(d.values(), dtype=float)  # efficiënt en direct

    # 2) Schaal detecteren (0–1 of 0–100)
    if percentage_scale == "auto":
        vmin = float(np.nanmin(vals)) if vals.size > 0 else 0.0
        vmax = float(np.nanmax(vals)) if vals.size > 0 else 1.0
        if 0.0 <= vmin and vmax <= 1.0:
            scale = "0-1"
        elif 0.0 <= vmin and vmax <= 100.0:
            scale = "0-100"
        else:
            # Fallback: neem 0–1 aan
            scale = "0-1"
    else:
        scale = percentage_scale

    # 3) Normaliseer eventueel naar 0–1
    if scale == "0-100":
        vals = vals / 100.0

    # 4) Range bepalen
    if value_range is None:
        value_range = (0.0, 1.0)

    # 5) Histogram tellen
    counts, bin_edges = np.histogram(vals, bins=bins, range=value_range)

    # 6) Plotten (optioneel)
    if save_path is not None or show:
        fig, ax = plt.subplots(figsize=(8, 4))
        ax.hist(vals, bins=bins, range=value_range, color="#4c78a8", edgecolor="white")
        ax.set_title("Histogram van percentages")
        # Labels: altijd in 0–1; wil je % tonen, pas hier aan (bijv. x-as in 0–100)
        ax.set_xlabel("Percentage (0–1)")
        ax.set_ylabel("Aantal rijen")
        ax.grid(alpha=0.2)
        plt.tight_layout()
        if save_path is not None:
            fig.savefig(save_path, dpi=150)
        if show:
            plt.show()
        plt.close(fig)
    
    return counts, bin_edges

```
